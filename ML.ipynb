## Loading the Dataset

```python
import pandas as pd

df = pd.read_csv("patients.csv",sep='\t')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Symptom'] = le.fit_transform(df['Symptom'])
```

```python
X = df.drop(['at_risk', 'stroke_risk', 'id'], axis=1)
y = df['at_risk']
```

## MODEL-1 Logistic Regression (Classification):

### Predict: Patient At Risk (0 or 1)

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X, y)

df['prediction'] = model.predict(X)
```

## MODEL-2 Linear Regression (Risk Score):

### Predict: Stroke Risk Value

```python
from sklearn.linear_model import LinearRegression

model2 = LinearRegression()
model2.fit(X, df['stroke_risk'])

df['risk_predicted'] = model2.predict(X)
```

## MODEL-3 K-Means Clustering:

### Group Patients into Categories

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
```

## Output:

```python
df.to_csv("predictions.csv", index=False)
```
